I'd like to see if we can identify a minimal network for C Elegans. Perhaps we can demonstrate that periphal neurons to the minimal network can add precision/nuance to the movement of the worm, but that the general dynamics can be approximated by a minimal network. 

It would be interesting to see if a theory of a minimal network is in fact supportable. This would enable us to show exactly what individual local processing units/peripheral functional pathways contribute. 

```python
connectome = dc.connectome_excel.data()
```

```python
# The sheetnames
connectome.keys()
```

```python
interneurons = connectome['Connectome']
muscle = connectome['NeuronsToMuscle']
sensory = connectome['Sensory']
```

```python
import networkx as nx
interneurons.shape
```

```python
%matplotlib inline
network = nx.DiGraph()

for row in xrange(interneurons.shape[0]):
    n_a = interneurons['Origin'][row]
    n_b = interneurons['Target'][row]
    num_connections = interneurons['Number of Connections'][row]
    neurotransmitter = interneurons['Neurotransmitter'][row]
    network.add_edge(n_a, n_b, 
         weight=num_connections, 
         neurotransmitter=neurotransmitter)
    
for row in xrange(muscle.shape[0]):
    n_a = muscle['Neuron'][row]
    n_b = muscle['Muscle'][row]
    num_connections = muscle['Number of Connections'][row]
    neurotransmitter = muscle['Neurotransmitter'][row]
    network.add_edge(n_a, n_b, 
         weight=num_connections, 
         neurotransmitter=neurotransmitter)
```

```python
network.edges('AIYR')
```

```python
sensory_neurons = sensory['Neuron']
```

Now that we've assembled a machine-readable connectome, I'd like to trace all forward facing pathways in the network. 

```python
not_visited = sensory_neurons
visited = {}
nodes = network.nodes()

while not_visited:
    node = nodes.pop(0)
    edges = network.edge(node)
    for s,e in edges:
        if not e in visited: 
            not_visited.append
```
